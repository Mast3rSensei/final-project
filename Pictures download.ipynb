```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import shutil
import urllib.request
```

```python
catalog = pd.read_csv('catalog.csv', encoding='latin9')
```

```python
catalog.isna().sum()
```

```python
p_catalog = catalog[catalog['FORM'] == 'painting']
```

```python
p_catalog.reset_index()
```

```python
p_catalog.shape
```

```python
# Look for potential rows which does not contain a valid URL
p_catalog['URL'].str.contains('^(?!.*(https://www.wga.hu)).*$', regex=True).sum()
```

```python
# Web-scrape and download all the images of paintings
paintings_url = []
headers = {'user-agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
for url in p_catalog['URL']:
    try:
        response = requests.get(url, headers=headers, timeout=0.2)
        html = response.content
        soup = BeautifulSoup(html, 'lxml')
        img_url = soup.find_all('a', href=True)
        img_url = (np.where(p_catalog['URL'] == url)[0][0], 'https://www.wga.hu' + img_url[1]['href'])
        paintings_url.append(img_url)
    except:
        print('Something went wrong')
        img_url = (np.where(p_catalog['URL'] == url)[0][0], 'Error')
        paintings_url.append(img_url)
```
