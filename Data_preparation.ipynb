```python
import pandas as pd
import numpy as np
import os
import random as rd
import re
```

```python
paintings_list = os.listdir('images')[1:]
train_sample = int(len(paintings_list)*0.8)
```

```python
train_imgs = rd.sample(paintings_list, train_sample)
evaluation_imgs = [painting for painting in paintings_list if painting not in train_imgs]
print('%d' % (len(train_imgs) + len(evaluation_imgs)))
```

```python
len(paintings_list)
```

```python
img_dir = 'images'
main_dir = 'ML_dataset'
train_dir = os.path.join(main_dir, 'train')
evaluation_dir = os.path.join(main_dir, 'evaluation')
for img in train_imgs:
    os.rename(os.path.join(img_dir, img), os.path.join(train_dir,  img))
for img in evaluation_imgs:
    os.rename(os.path.join(img_dir, img), os.path.join(evaluation_dir, img))
```

```python
paintings = pd.read_csv('catalog.csv', encoding='latin9')
urls = pd.read_csv('paintings_url.csv', index_col=0)
```

```python
paintings = paintings[paintings['FORM'] == 'painting'].reset_index()
```

```python
paintings.groupby('TYPE')['TITLE'].count()
```

```python
paintings['technique'] = [re.split('[,.:;]', technique)[0] for technique in paintings['TECHNIQUE']]
```

```python
paintings['technique'].unique()
```

```python
paintings.loc[paintings['technique'].str.contains('Oil (?!and)'), 'technique'] = 'Oil'
```

```python
paintings.loc[paintings['technique'].str.contains('(?=.*Tempera)(?!.*oil|Oil)(.+)'), 'technique'] = 'Tempera'
```

```python
paintings['TYPE'].unique()
```

```python
after_1400 = ['1401-1450',
 '1451-1500',
 '1501-1550',
 '1551-1600',
 '1601-1650',
 '1651-1700',
 '1701-1750',
 '1751-1800',
 '1801-1850',
 '1851-1900']
paintings.loc[~paintings['TIMELINE'].isin(after_1400), 'TIMELINE'] = 'prior 1400'
paintings['TIMELINE'].value_counts()
```

```python
train_indexes = []
for folder in os.listdir('ML_dataset/train/images'):
    for img in os.listdir('ML_dataset/train/images/' + folder):
        train_indexes.append(int(img.split('_')[0]))
train_indexes
```

```python
train_data_label = paintings.loc[train_indexes, 'TYPE']
train_data_label.to_csv('ML_dataset/train/labels/train_data_label.csv')
```

```python
evaluation_indexes = []
for folder in os.listdir('ML_dataset/evaluation/images'):
    for img in os.listdir('ML_dataset/evaluation/images/' + folder):
        evaluation_indexes.append(int(img.split('_')[0]))
evaluation_indexes
```

```python
evaluation_data_label = paintings.loc[evaluation_indexes, 'TYPE']
evaluation_data_label.to_csv('ML_dataset/evaluation/labels/evaluation_data_label.csv')
```
